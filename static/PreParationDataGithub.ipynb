**mengonversi nilai dalam kolom genre_lagu  menjadi list Python asli**




```python
import ast

def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i)
  return L

data['genre_lagu'] = data['genre_lagu'].apply(convert)
```

**Menghapus Spasi**

```python
data['genre_lagu'] = data['genre_lagu'].apply(lambda x:[i.replace(" ", "") for i in x])
data['nama_artis'] = data['nama_artis'].apply(lambda x:x.replace(" ", ""))

data.head()
```

**memisahkan per kata**

```python
data['Lirik_lagu'] = data['Lirik_lagu'].apply(lambda x:x.split())
data['nama_artis'] = data['nama_artis'].apply(lambda x:x.split())
data.head()
```

**menggabungkan kolom lirik lagu, nama artis dan genre**

```python
data['Label'] = data['Lirik_lagu'] + data['nama_artis'] + data['genre_lagu']
data.head(10)
```

**membuat variable data baru untuk memuat data baru yang diperlukan**

```python
new_data_github = data[['uri', 'artis_judulLagu', 'Label']]
```

**menggabungkan elemen-elemen dalam list pada kolom Label menjadi satu string**

```python
new_data_github['Label'] = new_data_github['Label'].apply(lambda x:" ".join(x))
```

```python
new_data_github['Label'][0]
```

**melakukan stemming pada teks, yaitu mengubah setiap kata dalam teks ke bentuk dasar (root word) menggunakan algoritma Porter Stemmer dari NLTK.**

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

**proses stemming pada setiap teks dalam kolom Label di DataFrame new_data_indo, sehingga semua kata dalam teks diubah ke bentuk dasarnya**

```python
new_data_github['Label'] = new_data_github['Label'].apply(stem)
```

```python
new_data_github['Label'][0]
```

**mengubah semua teks dalam kolom Label di DataFrame new_data_indo menjadi huruf kecil (lowercase), sehingga lebih konsisten untuk analisis teks**

```python
new_data_github['Label'] = new_data_github['Label'].apply(lambda x:x.lower())
```
